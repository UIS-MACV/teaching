# Thresholding & Transformaciones generales espaciales.

Cuando hablamos de Thresholding hacemos referencia a los métodos o algorítmos de valor umbral (umbralización),
cuya finalidad es segmentar objetos de una imagen que nos interesan del resto.
En las situaciones más sencillas se puede decidir que píxeles conforman los objetos que buscamos y que píxeles son sólo el entorno de estos objetos. 

Algorítmos básicos:
- Binary Thresholding
- Inverse Binary Thresholding
- Truncate Thresholding
- Threshold to zero
- Inverted Threshold to Zero
- Thresholding RGB

# Binary Thresholding


### Pseudocódigo:
if src(x,y) > threshold  
  dst(x,y) = maxValue  
else  
  dst(x,y) = 0  

Donde $src$ es la imagen de entrada, $dst$ es la imagen de sálida y threshold es el valor umbral

```python
# import opencv
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Read image
src = cv2.imread('numbers.jpg',0)

fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src, cmap = 'gray', interpolation = 'bicubic')
fig.add_subplot(122);
plt.hist(src.flatten());
```

```python
src = cv2.imread('numbers.jpg',0)
# Define Threshold
thresh = 12 # Intensity in grayscale.
# Define maxValue
maxValue = 255
```

```python
dst = src
for x in range(src.shape[0]):
    for y in range(src.shape[1]):
        if src[x,y] > thresh:
            dst[x,y] = maxValue
        else:
            dst[x,y] = 0      
```

```python
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src, cmap = 'gray', interpolation = 'bicubic')
fig.add_subplot(122);
plt.hist(src.flatten());
```

## Binary thresholding usando OpenCV

```python
# Read image
src = cv2.imread("numbers.jpg", cv2.IMREAD_GRAYSCALE)
 
# Set threshold and maxValue
thresh = 12
maxValue = 255
 
# Basic threshold example
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY);
```

```python
plt.imshow(dst, cmap='gray', interpolation='bicubic');
```

# Inverse Binary Thresholding

if src(x,y) > thresh  
  dst(x,y) = 0

```python
src = cv2.imread('numbers.jpg',0)
# Define Threshold
thresh = 12 # Intensity in grayscale.
# Define maxValue
maxValue = 255
```

## Inverse Binary thresholding usando OpenCV

```python
src = cv2.imread('numbers.jpg',0)
# Define Threshold
thresh = 12 # Intensity in grayscale.
# Define maxValue
maxValue = 255

#Opencv function
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY_INV)

# plot
plt.imshow(dst, cmap='gray', interpolation='bicubic');
```

# Truncate Thresholding

### Pseudocódigo

if src(x,y) > thresh  
  dst(x,y) = thresh  
else  
  dst(x,y) = src(x,y)  

```python
src = cv2.imread('numbers.jpg',0)
# Define Threshold
thresh = 12 # Intensity in grayscale.
# Define maxValue
maxValue = 255
```

```python
dst = src[:,:]
for x in range(src.shape[0]):
    for y in range(src.shape[1]):
        if src[x,y] > thresh:
            dst[x,y] = thresh
        else:
            dst[x,y] = src[x,y] 
```

```python
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic')
fig.add_subplot(122);
plt.hist(dst.flatten());
```

## Trunc Binary thresholding usando OpenCV

```python
src = cv2.imread('numbers.jpg',0)
# Define Threshold
thresh = 128 # Intensity in grayscale.
# Define maxValue
maxValue = 255

#Opencv function
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TRUNC)

# plot
plt.imshow(dst, cmap='gray');
```

Aplicando threshold a un solo canal de color

```python
src = cv2.imread('numberscolors.jpg',1)
dst = src[:,:,:]
for r in range(src.shape[0]):
    for c in range(src.shape[1]):
        if src[r,c,0] > 20:
            dst[r,c,0] = 255
        else:
            dst[r,c,0] = 0
plt.imshow(dst);
```

Extrayendo números dependiendo del color

```python
src = cv2.imread('numberscolors.jpg',1)
boundaries = [
    ([20, 100, 100], [30, 255, 255]),
    ([100, 100, 100], [120, 255, 255]),
    ([25, 146, 190], [62, 174, 250]),
    ([45, 12, 50], [245, 233, 228]),
    ([0, 0, 0], [128, 128, 128]),
    ([10, 200, 200], [160, 220, 255]),
    ([10, 8, 6], [145, 100, 90]),
    ([20, 20, 100], [145, 133, 128]),
    ([30, 40, 50], [165, 173, 228])
]
```

```python
# loop over the boundaries
dsts = []
for (lower, upper) in boundaries:
    # create NumPy arrays from the boundaries
    lower = np.array(lower, dtype = "uint8")
    upper = np.array(upper, dtype = "uint8")
 
    # find the colors within the specified boundaries and apply
    # the mask
    mask = cv2.inRange(src, lower, upper)
    output = cv2.bitwise_and(src, src, mask = mask)
    dsts.append(output)
```

```python
plt.figure(figsize=(40, 43))
for i in range(1,len(dsts)+1):
    ax = plt.subplot(5,5,i)
    plt.imshow(dsts[i-1])
plt.show()
```

# Transformaciones espaciales generales

- Traslaciones
- Redimensionado
- Rotaciones
- Transformación afin
- Transformación en perspectiva

```python
import cv2
import numpy as np
```

## Traslación de una imagen

```python
src = cv2.imread('deportista.jpg')
rows,cols = src.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(src,M,(cols,rows))

# plot
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src)
fig.add_subplot(122);
plt.imshow(dst);
```

# Redimensionamiento de una imagen

```python
src = cv2.imread('deportista.jpg')
height, width = src.shape[:2]
dst = cv2.resize(src,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

# plot
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src)
fig.add_subplot(122);
plt.imshow(dst);
```

## Rotación de una imagen

```python
src = cv2.imread('deportista.jpg')
rows,cols = src.shape[:2]

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(src,M,(cols,rows))

# plot
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src)
fig.add_subplot(122);
plt.imshow(dst);
```

## Transformación afin de una imagen

```python
src = cv2.imread('deportista.jpg')
rows,cols,ch = src.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(src,M,(cols,rows))

# plot
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src)
fig.add_subplot(122);
plt.imshow(dst);
```

## Transformación en perspectiva de una imagen

```python
src = cv2.imread('deportista.jpg')
rows,cols,ch = src.shape

pts1 = np.float32([[56,65],[368,52],[28,300],[389,390]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(src,M,(300,300))

# plot
fig = plt.figure(figsize=(15,5))
fig.add_subplot(121); 
plt.imshow(src)
fig.add_subplot(122);
plt.imshow(dst);
```
